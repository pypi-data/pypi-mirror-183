```python
"""
This python script demonstrates the creation of plasmas
"""

import paramak

    
ITER_simple = paramak.Plasma(
    major_radius=620,
    minor_radius=210,
    triangularity=0.33,
    elongation=1.85,
    name="ITER_plasma",
    rotation_angle=1,
    color=(0,1,0)
)

ITER = paramak.PlasmaBoundaries(
    A=-0.155,
    major_radius=620,
    minor_radius=210,
    triangularity=0.33,
    elongation=1.85,
    name="ITER_plasma_plasmaboundaries",
    rotation_angle=1,
    color=(1,0,0),
    azimuth_placement_angle=(360/7)
)

EU_DEMO = paramak.Plasma(
    name="EU_DEMO_plasma",
    major_radius=910,
    minor_radius=290,
    triangularity=0.33,
    elongation=1.59,
    rotation_angle=1,
    color=(0.9, 0.1, 0.1),
    azimuth_placement_angle=(360/7)*2
)

ST = paramak.Plasma(
    name="ST_plasma",
    major_radius=170,
    minor_radius=129,
    triangularity=0.55,
    elongation=2.3,
    rotation_angle=1,
    color=(0.2, 0.6, 0.2),
    azimuth_placement_angle=(360/7)*3
)

AST = paramak.Plasma(
    name="AST_plasma",
    major_radius=170,
    minor_radius=129,
    triangularity=-0.55,
    elongation=2.3,
    rotation_angle=1,
    color=(0, 0, 0),
    azimuth_placement_angle=(360/7)*4
)

NSTX_double_null = paramak.PlasmaBoundaries(
    name="NSTX_double_null_plasma_plasmaboundaries",
    A=0,
    major_radius=850,
    minor_radius=680,
    triangularity=0.35,
    elongation=2,
    color=(1, 1, 0),
    rotation_angle=1,
    configuration="double-null",
    azimuth_placement_angle=(360/7)*5
)

NSTX_single_null = paramak.PlasmaBoundaries(
    name="NSTX_single_null_plasma_plasmaboundaries",
    A=-0.05,
    major_radius=850,
    minor_radius=680,
    triangularity=0.35,
    elongation=2,
    color=(0.6, 0.3, 0.6),
    rotation_angle=1,
    configuration="single-null",
    azimuth_placement_angle=(360/7)*6
)

# combines all the plasma into a single object
all_plasmas = paramak.Reactor([ITER_simple, ITER, EU_DEMO, ST, AST, NSTX_double_null, NSTX_single_null])

all_plasmas.show()    
```

```python
all_plasmas.export_html()
```
